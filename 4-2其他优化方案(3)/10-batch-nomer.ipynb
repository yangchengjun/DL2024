```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 设置随机种子以确保结果可重复
np.random.seed(42)

# 生成标准正态分布的数据,方差为1，均值为0
data = np.random.normal(loc=0, scale=1, size=1000)

# 创建图形和坐标轴
plt.figure(figsize=(10, 6))

# 绘制数据分布的直方图
sns.histplot(data, bins=30, kde=True, stat='density', color='blue', alpha=0.6)

# 添加图形标题和标签
plt.title('Standard Normal Distribution (Mean = 0, Variance = 1)', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Density', fontsize=14)

# 添加垂直线表示均值
plt.axvline(x=0, color='red', linestyle='--', label='Mean = 0')

# 添加图例
plt.legend()

# 显示图形
plt.show()
```
